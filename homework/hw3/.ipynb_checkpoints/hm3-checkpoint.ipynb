```python
passanger_amount = data_titanic['PassengerId'].max()
pclass = data_titanic.groupby('Pclass')["Survived"].sum().apply(lambda x: x / passanger_amount)
```

```python
for i in range(3):
    print('Class {} = {:.1f}%'.format(i+1, float(pclass.iloc[i]) * 100.0))
```

```python
#Функция для формирования списков из имён
def create_list_from_group(list_len, group):
    _list = []
    for i in range(list_len):
        name = group.iloc[i]
        bracket_index = name.find('(')
        if bracket_index != -1:
            space_index = name[bracket_index + 1:].find(' ')
            _list.append(name[bracket_index + 1: bracket_index + space_index + 1])
        else:
            dot_index = name.find('.')
            space_index = name[dot_index + 2:].find(' ')
            if space_index != -1:
                _list.append(name[dot_index + 2: dot_index + space_index + 2])
            else:
                _list.append(name[dot_index + 2:])
    return _list
```

```python
names_by_sex = data_titanic.groupby('Sex')['Name']
print('Самые популярные имена:')
for sex in ['male', 'female']:
    names_list = create_list_from_group(int(names_by_sex.size()[sex]), names_by_sex.get_group(sex))
    print('{} - {}'.format(sex, pd.Series(names_list).value_counts().idxmax()))
```

```python
names_by_class = data_titanic.groupby(['Sex', 'Pclass'])['Name']
print('Самые популярные имена:')
for sex in ['male', 'female']:
    for pclass in range(data_titanic['Pclass'].max()):
        names_list = create_list_from_group(int(names_by_class.size()[sex, pclass + 1]), names_by_class.get_group((sex, pclass + 1)))
        print('{}, class {} - {}'.format(sex, pclass + 1, pd.Series(names_list).value_counts().idxmax()))
```
