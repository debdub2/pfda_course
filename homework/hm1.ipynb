```python
class Chocolate:

  def __init__(self, price, additive = None):
    self.additive = additive
    self.price = float(price)
    if self.price <= 0:
      raise ValueError('Only positive numbers are allowed')


  def __str__(self):
    return f"Шоколадный батончик, {self.price} рублей"


  def show_my_chocolate(self):
    if self.additive is None:
      print('Обычный шоколадный батончик')
    else:
      print(f'Шоколадный батончик, добавка: {self.additive}')


def main():
  simple_choc = Chocolate(87.75)
  simple_choc.show_my_chocolate()
  print(simple_choc)
  add_choc = Chocolate(92, 'Ваниль')
  add_choc.show_my_chocolate()
  print(add_choc)
  negative_choc = Chocolate(-78, 'Шоколад?')


main()
```
